```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline

with open('P000-plot.txt', 'r') as f:
    lines = f.readlines()
    x = [float(line.split()[4]) for line in lines]
    y = [float(line.split()[1]) for line in lines]
    
plt.semilogx(x, y) 
plt.gca().invert_xaxis()
#plt.xscale("log")
#plt.xlabel('log $\kappa$')
plt.ylabel('Temperature (K)')
plt.xlabel('$\kappa$')
plt.savefig('T-kappa.jpeg', dpi=1000)
plt.show() 
```

```python
%matplotlib inline

with open('P000-plot.txt', 'r') as f:
    lines = f.readlines()
    x = [float(line.split()[4]) for line in lines]
    y1 = [float(line.split()[2]) for line in lines]
    y2 = [float(line.split()[5]) for line in lines]
    
plt.loglog(x, y1, label='P$_{gas}$') 
plt.loglog(x, y2, label='P$_{rad}$') 
#plt.xscale("log")
#plt.yscale("log")
plt.gca().invert_xaxis()
plt.xlabel('$\kappa$')
plt.ylabel('log P (dyne cm$^{-2}$)')
plt.legend()
plt.savefig('P-kappa.jpeg', dpi=1000)
plt.show() 
```
