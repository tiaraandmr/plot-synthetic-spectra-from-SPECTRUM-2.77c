```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline

with open('P000-ir-2.spc', 'r') as f:
    lines = f.readlines()
    x = [float(line.split()[0]) for line in lines]
    y = [float(line.split()[1]) for line in lines]
    
with open('P002-ir-2.spc', 'r') as f:
    lines = f.readlines()
    x1 = [float(line.split()[0]) for line in lines]
    y1 = [float(line.split()[1]) for line in lines]
plt.ylim(0,1)

plt.rcParams['font.size'] = '20'

# plotting a line plot with it's default size 
print("Plot in it's default size: ") 
plt.plot(x, y) 
plt.show() 
  
# plotting a line plot after changing it's width and height 
f = plt.figure() 
f.set_figwidth(25) 
f.set_figheight(4) 

y1 = np.array(y1) 
offset = -0.3
  
print("Plot after re-sizing: ") 
plt.plot(x, y, color='#741b47', label='[Fe/H] = 0, log g = 2, $\u03BE = 0$') 
plt.plot(x, y1, color='#f1c232', label='[Fe/H] = 0, log g = 2, $\u03BE = 2$') 
plt.ylim(0.3,1.1)
plt.xlim(15650,16050)
plt.legend()
plt.savefig('P000-P002_1.jpeg', dpi=1200)
plt.show() 

# plotting a line plot after changing it's width and height 
f = plt.figure() 
f.set_figwidth(25) 
f.set_figheight(4) 

plt.plot(x, y, color='#741b47', label='[Fe/H] = 0, log g = 2, $\u03BE = 0$') 
plt.plot(x, y1, color='#f1c232', label='[Fe/H] = 0, log g = 2, $\u03BE = 2$') 
plt.ylim(0.3,1.1)
plt.xlim(16950,17350)
plt.savefig('P000-P002_2.jpeg', dpi=1200)
#plt.legend()
plt.show() 

# plotting a line plot after changing it's width and height 
f = plt.figure() 
f.set_figwidth(25) 
f.set_figheight(4) 

plt.plot(x, y, color='#741b47', label='[Fe/H] = 0, log g = 2, $\u03BE = 0$') 
plt.plot(x, y1, color='#f1c232', label='[Fe/H] = 0, log g = 2, $\u03BE = 2$')
plt.ylim(0.3,1.1)
plt.xlim(22750,23800)
plt.savefig('P000-P002_3.jpeg', dpi=1200)
plt.show() 
```
